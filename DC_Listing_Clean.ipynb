```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dc_listings_data = pd.read_csv("DC_Listings_csv.csv")

dc_listings_data.columns
```

```python
dc_listings_data = dc_listings_data.set_index('id', drop = False)
```

```python
#dc_listings_data.head()
dc_listings_data1 = dc_listings_data[['id','host_listings_count','host_response_rate','host_acceptance_rate','accommodates','bedrooms', 'bathrooms', 'beds', 'price','security_deposit', 'cleaning_fee','minimum_nights'
                                      , 'maximum_nights','number_of_reviews','neighbourhood_cleansed','latitude', 'longitude']]
```

```python
dc_listings_data1.head()
```

```python
dc_listings_data1.shape
```

```python
### Data cleansing complete
#dc_listings_data1.head()
dc_listings_data1.to_csv('dc_listings_indexed.csv')
```

```python
dc_listings = dc_listings_data[['host_listings_count','host_response_rate','host_acceptance_rate'
,'accommodates','property_type','room_type','bedrooms','bathrooms','beds','bed_type'
,'price','security_deposit','cleaning_fee','minimum_nights','maximum_nights','number_of_reviews'
,'latitude','longitude','zipcode','city']]
```

```python
dc_listings.head()
```

```python
dc_listings = dc_listings[['accommodates','bedrooms', 'bathrooms', 'beds', 'price','security_deposit', 'cleaning_fee','minimum_nights', 'maximum_nights','number_of_reviews']]
dc_listings.dropna() 
dc_listings.fillna(0)
dc_listings.shape

```

```python
dc_listings.head()
```

```python
dc_listings['price'] = dc_listings['price'].str.replace("\$|,",'')
dc_listings['price'] = dc_listings['price'].fillna(0).astype(float)
dc_listings['cleaning_fee'] = dc_listings['cleaning_fee'].str.replace("\$|,",'')
dc_listings['cleaning_fee'] = dc_listings['cleaning_fee'].fillna(0).astype(float)
dc_listings['security_deposit'] = dc_listings['security_deposit'].str.replace("\$|,",'')
dc_listings['security_deposit'] = dc_listings['security_deposit'].fillna(0).astype(float)
dc_listings['bedrooms']         = dc_listings['bedrooms'].fillna(0).astype(float)
dc_listings['bathrooms']        = dc_listings['bathrooms'].fillna(0).astype(float)
dc_listings['beds']             = dc_listings['beds'].fillna(0).astype(float)
```

```python
dc_listings.head()
```

```python
### Data cleansing complete
dc_listings.to_csv('dc_listings_clean.csv')
```
